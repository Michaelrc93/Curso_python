```python
# Siempre lo primero que debo hacer es importar pandas y numpy
import pandas as pd
import numpy as np
```

```python
#Ahora subimos el archivo desde google drive, ya no debemos logearnos de nuevo
df_estudiantes=pd.read_csv('/content/drive/MyDrive/StudentsPerformance (1).csv')
df_estudiantes
```

```python
#Sigamos con la creaciÃ³n de columnas usando el metodo assign o el mÃ©todo insert
#Si quiero agregar varias columnas uso assign y tambien si quiero sobreescribir una columna
```

```python
#Fisica y filosofia aleatorias y las guardamos en variables
filosofia_scrore=np.random.randint(1, 100, size=1000)
fisica_score=np.random.randint(1, 100, size=1000)
```

```python
#Ahora le pondre un indice a cada fila para que sepa en donde va el dataframe
filosofia_nota=pd.Series(filosofia_scrore, index=np.arange(0,1000))
fisica_nota=pd.Series(fisica_score, index=np.arange (0,1000))
```

```python
#Ahora lo qe voy hacer es poner las nuevas columnas a mi dataframe, usando assign
df_estudiantes=df_estudiantes.assign(filosofia_scrore=filosofia_nota,fisica_score=fisica_nota)
```

```python
#Ahora vamos a usar el metodo insert, el metodo insert es diferente por que me permite especificar donde yo voy a colocar esa columna

# Al usar el metodo insert debemos tener cuidado po que no me devuelve copias si no que el cambia
df_estudiantes.insert(1, 'EDU_SCORE', fisica_nota)
```
