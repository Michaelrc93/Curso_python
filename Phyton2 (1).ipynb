Para poder usar dataframe vamos activar dos bibliotecas una de ellas es numpy y la otra es pandas, las dos se complementan y hacen cosas diferentes para crear dataframe existen varias posibilidades sin embargo la mas usada es la lectura desde un archivo

**Opcion 1**

```python
import pandas as pd # importo pandas y lo nombro pd
import numpy as np # importo numpy y lo nombro np
```

```python
# la primera forma de crear un dataframe es usando el metodo array y luego guardarlo en una variable
dataframe_1= np.array([[2,4],[6,8],[10,12],[14,16]])
```

```python
# Para crear el dataframe del arreglo (array) anterior debemos usar pandas
pd.DataFrame(dataframe_1)
```

```python
dataframe_2=np.array([[5,10,15],[20,25,30],[35,30,45],[50,55,60]])
pd.DataFrame(dataframe_2)
```

```python
dataframe_3=np.array([[3,6,9],[12,15,18],[21,24,27],[30,33,36]])
pd.DataFrame(dataframe_3)
```

```python
# Ahora vamos a cambiarle el nombre a las columnas eso se hace de la siguiente manera:
df1=pd.DataFrame(dataframe_1,index=["Fila1","Fila2","fila3","fila4"],columns=["Columna1","columna2"])
df1
```

```python
# Ahora se cambiara el nombre a las columnas y filas
df3=pd.DataFrame(dataframe_3,index=["Fila1","Fila2","Fila3","Fila4"],columns=["Columna1","Columna2","Columna3"])
df3
```

```python
# Ahora creamos un dataframe usando listas
dataframe_4=[[2,4],[6,8],[10,12]]
```

```python
df4=pd.DataFrame(dataframe_4,index=["fila1","Fila2","Fila3"],columns=["Columna1","Columna2"])
df4
```

```python
dataframe_5=[["Michael","Rodriguez",31],["alvaro","Rodriguez",50],["Juliana","Silva",30],["maria","Camargo",48]]
df5=pd.DataFrame(dataframe_5,index=["Usuario1","Usuario2","Usuario3","Usuario4"],columns=["Nombre","Apellido","Edad"])
df5
```

Para el tercer metodo uso los diccionarios de la siguiente manera:

```python
#Primero creamos dos listas, puedo crear cuantas listas quiera, entiendo que las listas son las columnas del dataframe
CiudadColombia=["Bogota","Medellin","Cali","Barranquilla"]
CiudadesInfo=[11500,2500,2200,1300]
```

```python
#Ahora creamos los diccionarios con las listas
Ciudades={'Ciudades':CiudadColombia,'Poblacion MM':CiudadesInfo}
df6=pd.DataFrame(Ciudades)
```

```python
#Llamo el dataframe
df6
```

```python
# cuatro paises, poblacion, expectativa de vida
Paises=["Colombia","Mexico","China","Japon"]
poblacion=[51000,126000,1410000,121561]
expectativa_de_vida=[77,74,78,87]
```

```python
#Ahora creamos los diccionarios con las listas
Info_Paises={'Pais':Paises,'Poblacion en MM':poblacion,'Expectativa de vida en a√±os':expectativa_de_vida}
df7=pd.DataFrame(Info_Paises)
```
